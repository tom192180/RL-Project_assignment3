```python
import pickle
import matplotlib.pyplot as plt 
import numpy as np
```

```python
with open('cartpole_NDQN_final.pickle', 'rb') as handle:
    NDQN = pickle.load(handle)
```

```python
with open('cartpole_DDQN_final.pickle', 'rb') as handle:
    DDQN = pickle.load(handle)
```

```python
with open('cartpole_DuelingDDQN_final.pickle', 'rb') as handle:
    DuelingDDQN = pickle.load(handle)
```

```python
with open('cartpole_DuelingDDQN_final_discre.pickle', 'rb') as handle:
    DuelingDDQN_dist = pickle.load(handle)
```

```python
container = [NDQN,DDQN,DuelingDDQN,DuelingDDQN_dist]
for r_sums, name in zip(container,["NDQN","DDQN","DuelingDDQN","DuelingDDQN_dist"]):
    for i in range(len(r_sums)):
        if r_sums[i] == 200:
            break
    print(name, "total timesteps consumed: ",r_sums[-1] )
    print("first reach 200",i)
    print("R_sums",np.round(np.mean(np.array(r_sums[i:-1])), 3),"±", 2*np.round(np.std(np.array(r_sums[i:-1])), 3))
    print()
```

```python
#NDQN
plt.plot(NDQN[:-1])
plt.xlabel("Episode Number")
plt.ylabel("Reward")
```
